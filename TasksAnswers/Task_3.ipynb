# Loop

1. Print numbers from 1 to 10 using a for loop.


```python
for i in range(1, 11) :
    print(i, end = ' ')
```

2. Print all even numbers from 1 to 20.


```python
for i in range(1,21):
    if i % 2 == 0 :
        print(i, end = ' ')
```

3. Print each character of the string: 'Python'.


```python
text = 'Python'
for i in text :
    print(i, end = ' ')
```

4. Using a while loop, print numbers from 5 down to 1.

```python
n = 5
while n > 0 :
    print(n, end = ' ')
    n -= 1
```


5. Write a loop to find the sum of numbers from 1 to 50.

```python
sum = 0
for i in range(1, 51) :
    sum += i
print(sum)
```


6. Print the multiplication table of 5 (from 5×1 to 5×10).

```python
for i in range(1, 11) :
    print(f"{i} * 5 = {i*5}")
```


7. Find how many vowels are present in the string 'Programming'.

```python
text = 'Programming'
vowelCount = 0
for i in text.lower() :
    if i in 'aeiou' :
        vowelCount += 1
print('Number of vowels in Programming : ',vowelCount)
```


8. Use a loop to reverse the string 'PythonLoops'.

```python
text = 'PythonLoops'
revText = ''
for i in text :
    revText = i + revText
print(revText)
```


9. Print numbers from 1–10, but skip 5 using continue.

```python
for i in range(1, 11) :
    if i == 5 :
        continue
    print(i, end = ' ')
```


10. Print numbers from 1–20, but stop when number reaches 13 using break.

```python
for i in range(1, 21) :
    if i == 13 :
        break
    print(i, end = ' ')
```


11. Write a loop to check if a number is prime.

```python
import math
n = int(input("Enter a number : "))
prime = True
for i in range(2, int(math.sqrt(n))+1) :
    if n % i == 0 :
        prime = False
print(n, "is prime" if prime else "is not prime")
```


12. Count how many times each character occurs in 'mississippi'.

```python
text = 'mississippi'
d = {}
for i in text :
    if i in d :
        d[i] += 1
    else :
        d[i] = 1
for i,j in d.items() :
    print(f"{i} : {j}")
```


13. Using nested loops, print the pattern: *, **, ***, ****, *****.

```python
for i in range(1, 6) :
    print(i*'*')
```


14. Write a loop to find the largest digit in the number 5847361.

```python
number = 5847361
maxdigit = 0
while number :
    maxdigit = max(number%10, maxdigit)
    number //= 10
print(maxdigit)
```

15. Write a loop to print the pattern: *****, ****, ***, **, *

```python
for i in range(5, 0, -1) :
    print(i * '*')
```
