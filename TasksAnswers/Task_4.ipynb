
2. Write a function greet(name) that prints 'Hello '.

```python
def greet(name) :
    print('Hello',name)
greet('Ravi')
```


3. Write a function square(n) that returns the square of n.

```python
square = lambda x : x * x
n = int(input('Enter a number :'))
print(f"Square of {n} is {square(n)}")
```


4. Write a function calculator(a, b) that returns the sum, difference, and product of a & b.

```python
def calculator(a, b) :
    sum = a + b
    diff = a - b
    prod = a * b
    print('Sum :', sum)
    print('Difference :', diff)
    print('Product :', prod)
calculator(6, 4)
```


5. Write a function country(name='India') that prints 'I am from '.

```python
def country(name='India') :
    print('I am from',name)
country()
country('America')
```


6. Write a function total(*nums) that returns the sum of all numbers passed.

```python
def total(*nums) :
    sum = 0
    for i in nums :
        sum += i
    return sum
print("Sum of all numbers :",total(1, 2, 3, 4, 5))
```


7. Write a function student_info(**data) that prints key : value for all items.

```python
def data(**data) :
    for i,j in data.items() :
        print(f"{i} : {j}") 
data(Name = 'Ravi', Age = 19)
```


8. Write a function count_vowels(text) that returns the number of vowels in the string.

```python
def count_vowels(text) :
    vowelsCount = 0
    for i in text :
        if i in "aeioAEIOU" :
            vowelsCount += 1
    return vowelsCount
print('Number of vowels in "Hello World" :',count_vowels('Hello World'))
```


9. Write a lambda function to return the cube of a number.

```python
cube = lambda x : x * x * x
n = 6
print(f"Cube of {n} is {cube(n)}")
```


10. Write a function unique_letters(text) that returns unique letters in the order they appear

```python
def unique_letters(text) :
    unique = []
    for i in text :
        if text.count(i) == 1 :
            unique.append(i)
    return unique
text = "Welcome to Programming"
print(*unique_letters(text))
```
