## SKILL FREQUENCY CHECKER
### Objective :
### - Take a paragraph of resume text.
### - Take a skill from the user.
### - Count how many times the skill appears.
### - Print: “Skill appears X times.”

```python
paragraph = input("Enter the paragraph :")
skill = input("Enter a skill to count frequency :").lower()
print("Given Paragraph :\n",paragraph)
print("Entered skill :",skill)
```

```python
count = 0 # or we can also use count = sum(1 for word in words if word==skill)
words = [word.strip("!.,;:/-()") for word in paragraph.lower().split()]
for word in words :
    if skill == word :
        count += 1
print(skill.title(),"appears",count,'times')
```
