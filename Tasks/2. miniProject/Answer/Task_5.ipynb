##  SIMPLE RESUME CATEGORY CLASSIFIER
### Objective :
### - Input a list of skills.
### - Use simple if/elif logic to classify:
###     . Data/ML
###     . Web Development
###     . Software Development
###     . Other
### - Print the predicted category.


```python
categories = {
    'Data/ML' : 'python, machine learning algorithms, data preprocessing, model training, model evaluation, numpy, pandas, scikit-learn, tensorflow/pytorch, statistics, data visualization, problem solving, data wrangling',

    'Web Development' : 'html, css, javascript, react/vue/angular, responsive design, node.js/django/flask, apis, debugging, version control, git, databases, sql, mongodb, frontend development, backend development',

    'Software Developer / Software Engineer' : 'object-oriented programming, oop, python, java, c++, javascript, problem solving, data structures & algorithms, debugging, version control, git, databases, sql, mongodb, apis, web development',

    'Data Analyst' : 'excel, sql, data cleaning, data visualization, tableau, power bi, python (pandas, numpy), statistics, dashboards, reporting, critical thinking, business communication',

    'Graphic Designer' : 'adobe photoshop, illustrator, indesign, creativity, typography, branding, layout design, color theory, ui/ux basics, visual communication',

    'Cybersecurity Analyst' : 'networking fundamentals, linux, windows security, firewalls, intrusion detection, siem tools, penetration testing basics, threat analysis, incident response, risk assessment',

    'DevOps Engineer' : 'linux, ci/cd pipelines, docker, kubernetes, cloud platforms (aws, azure, gcp), scripting (bash/python), git, infrastructure as code (terraform), monitoring tools, automation',

    'UI/UX Designer' : 'figma, adobe xd, wireframing, prototyping, user research, design systems, visual design, usability testing, interaction design, creativity',

    'Product Manager' : 'roadmapping, market research, user research, communication, problem solving, agile/scrum, data analysis, prioritization, documentation, stakeholder management',

    'Cloud Engineer' : 'aws/azure/gcp, cloud architecture, linux, networking, virtualization, docker, kubernetes, scripting, cloud security, monitoring tools',

    'Business Analyst' : 'requirements gathering, sql, data analysis, documentation, communication, critical thinking, dashboards, process optimization, project coordination, stakeholder management',

    'Mobile App Developer' : 'java, kotlin, swift, react native, flutter, mobile ui design, debugging, apis, version control, git, testing',

    'Content Writer / Copywriter' : 'writing, storytelling, seo writing, grammar, research, creativity, editing, content planning, social media basics, communication',

    'IT Support Specialist' : 'hardware troubleshooting, software installation, networking basics, operating systems (windows, mac, linux), customer support, documentation, ticketing systems, problem solving'
}
```

```python
skills = input("Enter your skills seperated by comma(,) : ").split(',')
# input : Python, Java, C++, Data Structures & Algorithms, Object-Oriented Programming, SQL, MongoDB, REST APIs, Git, Debugging, Problem-Solving, Agile/Scrum, Cloud Basics (AWS/Azure)
cleanSkills = []
for skill in skills :
    clean = skill.strip().lower()
    cleanSkills.append(clean)
print("Your skills :", cleanSkills)
```

```python
predictedCategory = set()
# analying the predicted category
for skill in cleanSkills :
    for category,requiredSkills in categories.items() :
        if skill in requiredSkills :  # check if skill is in required skills so that we can add the category in set
            predictedCategory.add(category)

print("Predicated category :")
for c in predictedCategory :
    print(c)
```
