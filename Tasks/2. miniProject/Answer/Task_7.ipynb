## SIMPLE EMAIL EXTRACTOR (NO REGEX)
### Objective :
### - Input a text block.
### - Find the word containing “@”.
### - Print the extracted email.
### - If not found → print “No email detected.”

```python
paragraph = input("Enter a paragraph :")
print("Given paragraph :",paragraph)
```

```python
email = ''
if '@' in paragraph and '.com' in paragraph :
    for word in paragraph.split() :
        if '@' in word and '.com' in word :
            email = word.strip('. ')
    print("Email found :",email)
else :
    print("Email no found.")
```
