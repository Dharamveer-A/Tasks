## Resume word counter
### Objective :
### - Accept a paragraph of resume text.
### - Count total words.
### - Count unique words.
### - Identifiy the most repeated word.
### - Display the result.

```python
# accept a paragraph
import re
import collections as c
paragraph = input("Enter the paragaph :")
print("You typed paragraph :", paragraph)
totalWords = paragraph.split(' ') # Spliting all words
cleanedWords = []
for word in totalWords :
    clean = word.strip().lower() # Striping and convert it to lower case letters
    clean = re.sub(r'[^a-z0-9]','',clean)  # Cleaning each words
    clean = re.sub(r'\s+',' ',clean) # Replacing many spaces with one space
    cleanedWords.append(clean) # Appending clean to cleanedWords

frequency = c.Counter(cleanedWords) # Counting frequency with built-in function
# Counting the frequency each words
# frequency = dict()
# for word in cleanedWords :
#     if word in frequency :
#         frequency[word] += 1
#     else :
#         frequency[word] = 1
    
```

```python
# Counting total words
print("Total words in given paragraph :", len(frequency))
```

```python
# Unique
print("Uinque words :")
uniqueWords = []
for word,freq in frequency.items() :
    if freq == 1 :
        print(word)
        uniqueWords.append(word)
print("Count of unique words :", len(uniqueWords))
```

```python
print("Frequency of each word :")
for i,j in frequency.items() :
    print(f"{i} : {j}")
```

```python
sortedFrequency = sorted(frequency.items(), key = lambda x : x[1], reverse=True)
print("Word with most frequency : ",sortedFrequency[0][0], '->', sortedFrequency[0][1])
```

```python
# printing all the results 
print(f"Total words :{len(cleanedWords)}")
print(f"Total unique words :{len(uniqueWords)}")
print(f"Most repeated word is \"{sortedFrequency[0][0]}\" with frequency of {sortedFrequency[0][1]}")
```
