## Experience extractor
### Objective :
### -Input sentence like : "I have 3 years of experience in Python."
### -Extract the number of years using string methods.
### -Output : "Experience Deteceted : X years"

```python
# Getting the input as sentence
sentence = input("Enter the sentence") 
print("Your sentence :", sentence)
```

```python
# Extracting years of experience
years = ''
for word in sentence.split() :
    if word.isnumeric() and 'experience' in sentence:
        years = word

# Display the experience 
if years != '' :
    print(f"Experience Deteceted : {years} years")
else :
    print("Experience not Detected")
```
