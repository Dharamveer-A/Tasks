## JOB TITLE NORMALIZER
### Objective :
### -Input job titile text.
### -Remove special characters.
### -Convert text to title case.
### -print normailized job title.

```python
# Getting input as title
title = input("Enter job title :")
print("Job title :",title)
```

```python
# Removing special characters and making it to title case
import re
cleanedTitle = re.sub(r'[^a-zA-Z0-9 +.-]','',title)
cleanedTitle = re.sub(r'\s+',' ', cleanedTitle)
normalizedTitle = cleanedTitle.title()
print("Normalized Title :",normalizedTitle)
```
