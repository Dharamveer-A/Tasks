## TECH STACK COUNTER
### Objective :
### - Input a list of tools/technologies.
### - Program counts how many items belong to:
### . Programming Languages
### . Databases
### . Frameworks
### - Print a summary.

```python
technologies = {
    'Programming Languages': 'python, c, c++, java, javascript, html, css, sql, typescript',
    'Databases': 'mysql, mongodb, postgresql, sqlite, oracle, firebase, redis',
    'Frameworks': 'django, flask, fastapi, react, node.js, express.js, tensorflow, pytorch, keras, angular, spring boot'
}
```

```python
tools = input("Enter your tools seperated comma(,):")
print("Your tools :", tools)
```

```python
toolsList = [tool.strip('. ').lower() for tool in tools.split()]
summary = { 
    'Programming Languages' : set(),
    'Databases' : set(),
    'Frameworks' : set(),
    'Others' : set()
        }
for tool in toolsList :
    if tool in technologies["Programming Languages"]:
        summary["Programming Languages"].add(tool)
    elif tool in technologies["Databases"]:
        summary["Databases"].add(tool)
    elif tool in technologies["Frameworks"]:
        summary["Frameworks"].add(tool)
    else:
        summary["Others"].add(tool)
```

```python
print(f"Summary :\nProgramming Language :{len(summary["Programming Languages"])}\nDatabases :{len(summary["Databases"])}\nFrameworks :{len(summary["Frameworks"])}\nOthers :{len(summary["Others"])}")
```
