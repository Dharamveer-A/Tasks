## Basic skill matcher
### Objective :
####  - Take resume skills as comma seperated input.
####  - Take job required skill as input.
####  - Compare both list.
####  - Print matched and missing skills .

```python
#  Getting resume skills as input and cleaning it
import re 
resumeInput = input("Enter the resume skills : ")
resumeSkills = []
resumeSkills = [i.strip().lower() for i in resumeInput.split(',') if i not in resumeSkills]
# converting resumeInput as list and making each element as lower case ans striping it with strip() functions
# resumeSkills = set(list(map( str.lower, [skill.strip() for skill in input("Enter the resume skills seperated by comma(,) :").split(',')])))

cleanedResumeSkills = []
for skill in resumeSkills :
    clean = re.sub(r'[^a-z0-9 .#-+]','',skill) #[^a-z0-9 .#-+] if anything is not like this then it will replace with null character from skill string
    clean = re.sub(r'\s+',' ',clean) #Removes extra space by replacing it with single space 
    cleanedResumeSkills.append(clean) #Appending clean string in cleanedResumeSkills list
```

```python
print("Your resume skills :", cleanedResumeSkills)
```

```python
#  Getting job required skills as input and cleaning it
import re
jobRequiredInput = input("Enter the job requried skills : ")
jobRequiredSkills = []
jobRequiredSkills = [i.strip().lower() for i in jobRequiredInput.split(',') if i not in jobRequiredSkills]
# converting jobRequiredInput as list and making each element as lower case ans striping it with strip() functions

cleanedJobRequiredSkills = []
for skill in jobRequiredSkills :
    clean = re.sub(r'[^a-z0-9 .#-+]','',skill) #[^a-z0-9 .#-+] if anything is not like this then it will replace with null character from skill string
    clean = re.sub(r'\s+',' ',clean) #Removes extra space by replacing it with single space 
    cleanedJobRequiredSkills.append(clean) #Appending clean string in cleanedJobResumeSkills list
```

```python
print("Job required skills :", cleanedJobRequiredSkills)
```

```python
# Compare both and seperate matching and missing skills
matchingSkills = set(cleanedResumeSkills) & set(cleanedJobRequiredSkills) # Will give matching skills
missingSkills = set(cleanedJobRequiredSkills) - set(cleanedResumeSkills) # will give missing skills
print("Resume skills :", cleanedResumeSkills)
print("Job required skills :", cleanedJobRequiredSkills)
print("Matching skills : ", matchingSkills)
print("Missing skills : ", missingSkills)
```
