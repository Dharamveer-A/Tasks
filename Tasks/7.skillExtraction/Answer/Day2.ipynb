```python
import spacy
from spacy.matcher import PhraseMatcher

nlp = spacy.load("en_core_web_sm")
matcher = PhraseMatcher(nlp.vocab, attr="LOWER")

skills_list = ["Python", "SQL"]
patterns = [nlp(skill) for skill in skills_list]
matcher.add("SKILLS", patterns)

text = "I have experience in Python and SQL"
doc = nlp(text)

matches = matcher(doc)
for match_id, start, end in matches:
    print(doc[start:end].text)
```

6. Extract technical skills from the sentence: 'Experience in Python, NLP, and Machine Learning
with SQL.'

```python
import spacy
from spacy.matcher import PhraseMatcher

nlp = spacy.load("en_core_web_sm")
matcher = PhraseMatcher(nlp.vocab, attr="LOWER")

techinical_skills = ["Python", "SQL", "NLP", "Machine Learning"]
patterns = [nlp(skill) for skill in techinical_skills]
matcher.add("SKILLS", patterns)

sentence = "Experience in Python, NLP, and Machine Learning with SQL"

docs = nlp(sentence)
matches = matcher(docs)
for id, start, end in matches:
    print(docs[start:end].text)
```
