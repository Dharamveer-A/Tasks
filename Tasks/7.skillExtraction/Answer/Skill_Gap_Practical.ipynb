1. Remove duplicate skills from two given lists: resume skills and job description skills.

```python
resume_skills = ['python', 'docker', 'sql', 'ml', 'python', 'mongodb']
job_description_skills = ['python', 'pandas', 'ml', 'mongodb', 'flask', 'ml']
print(f"With duplicates :\nResume : {resume_skills}\nJob description: {job_description_skills}")
resume_skills = list(set(resume_skills))
job_description_skills = list(set(job_description_skills))
print(f"Without duplicates :\nResume : {resume_skills}\nJob description : {job_description_skills}")
```
