5. Load two text files (resume and job description) and confirm successful loading.

```python
import os
import pdfplumber as pp 
from docx import Document

resumeLocation = "files/resume1/resume_1.docx"
JDLocation = "files/jobDescription1.txt"

def getFileFormat(path) :
    return os.path.splitext(path)[1]

def load_file(path) :
    fileFormat = getFileFormat(path)
    if fileFormat == ".txt" :
        with open(path, "r", encoding = "utf-8") as file :
            return file.read()
    if fileFormat == ".pdf" :
        with pp.open(path) as file :
            content = ""
            for page in file.pages :
                content += page.extract_text()
            return content
    if fileFormat == ".docx" :
        document = Document(path)
        text = []
        for page in document.paragraphs :
            text.append(page.text)
        return "\n".join(text)
    
try :
    resumeContent = load_file(resumeLocation)
    print("Resume loaded successfully")
    JDContent = load_file(JDLocation)
    print("Jod Description loaded successfully")
except FileNotFoundError :
    print("One or Both file not found")
except Exception as e:
    print("Error :",e) 
```

10. Clean the extracted text after parsing and display both raw and cleaned versions.

```python
import os
import pdfplumber as pp 
from docx import Document
import re 

resumeLocation = "files/resume3/resume_3.docx"
JDLocation = "files/jobDescription3.txt"

def getFileFormat(path) :
    return os.path.splitext(path)[1]

def load_file(path) :
    fileFormat = getFileFormat(path)
    if fileFormat == ".txt" :
        with open(path, "r", encoding = "utf-8") as file :
            return file.read()
    if fileFormat == ".pdf" :
        with pp.open(path) as file :
            content = ""
            for page in file.pages :
                content += page.extract_text()
            return content
    if fileFormat == ".docx" :
        document = Document(path)
        text = []
        for page in document.paragraphs :
            text.append(page.text)
        return "\n".join(text)
    
def cleanText(content) :
    clean = re.sub(r"[^a-zA-Z0-9+.-]", "", content)
    clean = content.split()
    clean = " ".join(clean).lower()

    return clean 


try :
    resumeContent = load_file(resumeLocation)
    cleanResumeContent = cleanText(resumeContent)
    print("Resume loaded successfully")
    JDContent = load_file(JDLocation)
    cleanJDContent = cleanText(JDContent)
    print("Jod Description loaded successfully \n")
    print("Resume (raw):")
    print(resumeContent)
    print("Resume (clean):")
    print(cleanResumeContent+'\n')
    print("Job Description (raw) :")
    print(JDContent+'\n')
    print("Job Description (clean) :")
    print(cleanJDContent+'\n')
except FileNotFoundError :
    print("One or Both file not found")
except Exception as e:
    print("Error :",e) 
```
