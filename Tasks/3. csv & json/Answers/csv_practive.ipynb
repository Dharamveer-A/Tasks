1. Create a CSV file named students.csv with columns name, age, grade and write 3 student
records.

```python
import csv 
coloumnName = ['Name','age','grade']
with open("files/students.csv","w",newline='') as file :
    writer = csv.writer(file)
    writer.writerow(coloumnName)
    writer.writerow(['Aadi',19,9])
    writer.writerow(['Bharat',19,10])
    writer.writerow(['Ravi',19,8])
print("File writen successfully")
```

2. Write a Python program to read and print each row from employees.csv.

```python
import csv
with open("files/employees.csv","r") as file:
    reader = csv.reader(file)
    for row in reader :
        print(row)
```

4. Read a CSV using DictReader and print all key-value pairs for each row.

7. Read sales.csv and calculate total sales (item, price, quantity).

```python
import csv 
grandTotal = 0
with open("files/sales.csv","r") as file :
    reader = csv.DictReader(file)
    for row in reader :
        total = int(row["Quantity"]) * int(row["Price"])
        grandTotal += total
        print(f"{row["Product"]} -> {total}")
print(f"Total sales(GrandTotal) : {grandTotal}")
```

8. Convert a list of lists into a CSV file (id, name).

```python
import csv 
data = [["id", "name"], [101, "Elan"], [102, "Akilan"], [103, "Deva"]]
with open("files/details.csv","w") as file :
    writer = csv.writer(file)
    writer.writerows(data)
    print("Content written successfully.")
with open("files/details.csv", "r") as file :
    reader = csv.DictReader(file)
    print(f"id, name")
    for row in reader :
        print(f"{row["id"]}, {row["name"]}")
```

9. Filter and print rows from people.csv where age > 30.

```python
import csv 
with open("files/people.csv","r") as file :
    reader = csv.DictReader(file)
    print(f"Name, Age, City")
    for row in reader :
        if int(row["Age"]) > 30 :
            print(f"{row["Name"]}, {row["Age"]}, {row["City"]}")
```

10. Copy all contents from source.csv to copy.csv.

```python
import csv 
with open("files/source.csv","r") as source :
    reader = csv.reader(source)
    with open("files/copy.csv","w",newline='') as copy :
        writer = csv.writer(copy)
        writer.writerows(reader)
        print("Content writen successfully.")
with open("files/copy.csv","r") as file :
    reader = csv.reader(file)
    for row in reader :
        print(row)
```
