## JSON Practice Questions (10)

1. Convert the JSON string {'name': 'Alex', 'age': 22, 'city': 'Chennai'} into a Python dictionary.

```python
import json
data = '{"name":"Alex","age":22,"city":"Chennai"}'
dictData = json.loads(data)
print(dictData)
```

2. Convert the dictionary {'id': 101, 'product': 'Laptop', 'price': 55000} into a JSON string.

```python
import json
dictData = {
    'id' : 101,
    'product' : 'Laptop',
    'price' : 55000
    }
jsonData = json.dumps(dictData)
print(jsonData)
```

3. Read a JSON file named data.json and print all key-value pairs.

```python
import json 
with open("files/data.json",'r') as file :
    reader = json.load(file)
    for data in reader :
        print("id:",data["id"])
        print("name:",data["name"])
        print("city:",data["city"]) 
```

4. Write the dictionary {'name': 'Mia', 'age': 24, 'grade': 'A'} into a JSON file student.json.

```python
import json 
dictData = {
    'name' : 'Mia',
    'age' : 24,
    'grade' : 'A'
}
with open('files/student.json','w',newline='') as file :
    json.dump(dictData, file, indent=4) 
    print("Content written successfully.")
```

5. From JSON {'employee': {'name': 'John', 'age': 30, 'department': 'HR'}}, print only the department.

```python
import json 
data = '{"employee":{"name":"John","age":30,"department":"HR"}}'
jsonFile = json.loads(data)
print(jsonFile["employee"]["department"])
```

6. Update the dictionary {'name': 'Raj', 'age': 45} by changing age to 50 and convert it to JSON.

```python
import json
content = '{"name":"Raj","age":45}'
data = json.loads(content)
print("Before updating :",data)
data["age"] = 50
print("After updating :",data)
```

7. Read a JSON list ['red', 'green', 'blue'] and print the second item.

```python
import json
content = '[["red","green","blue"]]'
data = json.loads(content)
print(data[0][1])
```

8. From JSON {'users': [{'id':1,'name':'Alex'}, {'id':2,'name':'Mia'}]}, print all user names.

```python
import json
content = '{"users": [{"id":1,"name":"Alex"}, {"id":2,"name":"Mia"}]}'
data = json.loads(content)
for user in data["users"] :
    print(user["name"])
```

9. Add a new key 'country': 'India' to {'name': 'Sam', 'age': 28} and convert back to JSON.

```python
import json
content = {"name":"Sam","age":28}
print("Before adding :",content)
content["country"] = "India"
content = json.dumps(content, indent=4)
print("After adding :",content)
```

10. Create items.json with [{'item':'Pen','price':10}, {'item':'Book','price':50}, {'item':'Bag','price':700}]
and print total cost.
